# 1. 再帰型ニューラルネットワークの概念

## 1-1. RNN全体像
時系列データに対応可能な、ニューラルネットワーク
時系列データとは時系列順序に従って一定間隔で観測され、相互に統計的依存関係が認められるようなデータの系列である。<br>

RNNでは前の中間層からの出力（過去の情報）を利用する。<br>
RNNの特徴としては、前の中間層からの入力に掛ける重みがある。
<img src="./image/1-1_RNN.jpg">

### RNNの数学的記述
$$
u^t = W_{(in)} x^t + Wz^{t-1} + b\\
z^t = f(W_{(in)} x^t + Wz^{t-1} + b)\\
v^t = W_{(out)} z^t + c\\
y^t = g(W_{(out)} z^t + c)
$$

## 1-2. BTPP
$$
\frac{\partial E}{\partial W_{(in)}} = \frac{\partial E}{\partial u^t}\left[ \frac{\partial u^t}{\partial W_{(in)}} \right] = \delta^t \left[x\right]^t\\
\frac{\partial E}{\partial W_{(out)}} = \frac{\partial E}{\partial v^t}\left[ \frac{\partial v^t}{\partial W_{(out)}} \right] = \delta^{out,t} \left[z\right]^t\\
\frac{\partial E}{\partial W} = \frac{\partial E}{\partial u^t}\left[ \frac{\partial u^t}{\partial W} \right] = \delta^t \left[z\right]^{t-1}\\
\frac{\partial E}{\partial b} = \frac{\partial E}{\partial u^t}\frac{\partial u^t}{\partial b} = \delta^t\\
\frac{\partial E}{\partial c} = \frac{\partial E}{\partial v^t}\frac{\partial v^t}{\partial c} = \delta^{out,t}
$$
$\delta^t$と$\delta^{t-1}$に関して次の関係が成り立つ
$$
\delta^{t-1} = \frac{\partial E}{\partial u^{t-1}} = \frac{\partial E}{\partial u^t} \frac{\partial u^t}{\partial u^{t-1}}
= \delta^t \left[ \frac{\partial u^t}{\partial z^{t-1}} \frac{\partial z^{t-1}}{\partial u^{t-1}} \right]
= \delta^t \left[ Wf^{\prime}(u^{t-1}) \right]
$$
パラメータの更新式
$$
W_{(in)}^{t+1} = W_{(in)}^t - \epsilon \frac{\partial E}{\partial W_{(in)}} = W_{(in)}^t - \epsilon \sum={z=0}^{T_t} \delta^{t-z}[x^{t-z}]^T\\
W_{(out)}^{t+1} = W_{(out)}^t - \epsilon \frac{\partial E}{\partial W_{(out)}} = W_{(out)}^t - \epsilon \delta^{out,t}[z^t]^T\\
W^{t+1} = W^t - \epsilon \frac{\partial E}{\partial W} = W_{(in)}^t - \epsilon \sum={z=0}^{T_t} \delta^{t-z}[x^{t-z-1}]^T\\
b^{t+1} = b^t - \epsilon \frac{\partial E}{\partial b} = b^t - \epsilon \sum={z=0}^{T_t} \delta^{t-z}\\
c^{t+1} = c^t - \epsilon \frac{\partial E}{\partial c} = c^t - \epsilon \delta^{out,t}
$$
BTPPにおける誤差関数の更新
$$
\begin{eqnarray}
E^t &=& loss(y^t, d^t)\\
    &=& loss(g(W_{(out)}z^t+c), d^t)\\
    &=& loss(g(W_{(out)}f(W_{(in)}x^t+Wz^{t-1}+b)+c), d^t)
\end{eqnarray}
$$
ここでzに関して以下のように再帰的に展開できる
$$
W_{(in)}x^t+Wz^{t-1}+b\\
W_{(in)}x^t+Wf(u^{t-1})+b\\
W_{(in)}x^t+Wf(W_{(in)}x^{t-1}+Wz^{t-2}+b)+b
$$

### Q. 連鎖率の原理を使い、dz/dxを求めよ。
$$
z = t^2\\
t = x + y
$$

### A.
$$
\frac{dz}{dt} = 2t\\
\frac{dt}{dx} = 1\\
\frac{dz}{dx} = \frac{dz}{dt} \frac{dt}{dx} = 2(x+y)
$$
