# 5.Seq2Seq
Encoder-Decoderモデルの一つ

## 5-1. Encoder RNN
ユーザがインプットしたテキストデータを単語等のトークンに区切って渡す構造
<img src="./image/5_1_encoder_rnn.png">
Taking：文章を単語等のトークンごとに分割し、トークンのIDに分割する<br>
Embedding：IDからそのトークンを表す分散表現ベクトルに変換<br>
Encoder RNN：ベクトルを順番にRNNに入力していく<br>

<img src="./image/5_2_embedding.jpg">
embedding表現にすることで情報を圧縮している。<br>
単語の意味が似ているものは似たようなembedding表現になる。
### 処理手順
1つ目のベクトルをRNNに入力した隠れ層の出力を、2つ目のベクトルと一緒にRNNに入力する、といった処理を繰り返す。<br>
最後のベクトルを入力したときの隠れ層を保存する。

googleの自然言語処理モデルBERT

## 5-3. HRED
Seq2seqの課題
<br>
一問一答しかできない<br>
問に対して文脈がなく、応答が行われる<br>

- HRED
過去n-1 個の発話から次の発話を生成する。

- HREDとは
Seq2Seq+ Context RNN
<br>
Context RNN: Encoder のまとめた各文章の系列をまとめて、これまでの会話コンテキスト全体を表すベクトルに変換する構造

### 課題
- HRED は確率的な多様性が字面にしかなく、会話の「流れ」のような多様性が無い。
- 「うん」「そうだね」など短く情報量に乏しい答えをしがちである。

## 5-4. VHRED
HREDにVAEの潜在変数の概念を加えたもの

### オートエンコーダ構造説明
入力データから潜在変数zに変換するニューラルネットワークをEncoder<br>
逆に潜在変数zをインプットとして元画像を復元するニューラルネットワークをDecoderという
<img src="./image/5_4_auto_encoder.jpg">


## 5-5. VAE
通常のエンコーダでは、潜在変数zに何かしらデータを押し込めているものの、その構造がどのようなものかわからない<br>
VAEはこの潜在変数zに確率分布z～N(0,1)を仮定したもの<br>
データを潜在変数zの確率分布という構造に押し込めることを可能にする
