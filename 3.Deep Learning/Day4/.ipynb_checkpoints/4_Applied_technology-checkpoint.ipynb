# 4. 応用技術

## 4-1. MobileNets
https://qiita.com/HiromuMasuda0228/items/7dd0b764804d2aa199e4

### 論文タイトル
MobileNets: Efficient Convolutional Neural Networks for Mobile Vision Applications

- ディープラーニングモデルは精度は良いが、その分ネットワークが深くなり計算量が増える。
- 計算量が増えると、多くの計算リソースが必要で、お金がかかってしまう
- ディープラーニングモデルの軽量化・高速化・高精度化を実現(その名の通りモバイルなネットワーク)

#### 一般的な畳み込み層は計算量が多い
<img src="./image/4_1_general_convolution_layer.png">

### Depthwise Convolution
フィルタの数を1つで固定する。<br>
1つのチャネルに対し、1つのフィルタで畳み込み。<br>
入力と出力のサイズが同じ。
<img src="./image/4_2_depthwise_convolution_layer.png">

### Pointwise Convolution
カーネルのサイズを固定して、フィルタ数を変更する。<br>
出力がもとのサイズよりK×Kだけ小さくなる。
<img src="./image/4_3_pointwise_convolution_layer.png">

### MobileNetのアーキテクチャ
通常の畳み込みは空間方向とチャネル方向を同時に計算する。<br>
Depthwise Separable ConvolutionではDepthwise ConvolutionとPointwise Convolutionで別々に行う。<br>
計算量がDepthwise ConvolutionとPointwise Convolutionの和に削減できる。<br>
（MobileNetは今のところ第三世代まである？）

## BatchNorm
ミニバッチ単位での正規化。<br>
バッチサイズに影響を受ける。<br>
バッチの大きさをデバイス（計算資源）によって変えざるを得ないことがある。
（あまり使いたくない手法らしい）

## LayerNorm
全てのチャネルの平均と分散を求め正規化を実施<br>
ミニバッチの数に依存しない
https://www.slideshare.net/KeigoNishida/layer-normalizationnips

## Instance Norm
各sampleの各チャネルごとに正規化
