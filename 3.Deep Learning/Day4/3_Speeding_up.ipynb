# 3. 軽量化・高速化技術

### 分散深層学習とは

データ量の増大、パラメータ調整のために高速な計算が求められる<br>
複数の資源を使用し並列的にニューラルネットを構築し効率の良う学習したい<br>
データ並列化、モデル並列化、GPUによる高速化は不可欠<br>


## 3-1. モデル並列化
親モデルを各ワーカーに分割する。
<img src="./image/3_3_model_parallel_sync.png">
大きなモデルほど、並列化の効果が大きい。

## 3-2. データ並列
- モデルを各ワーカーに渡す
- データを分割し、ワーカーごとに計算させる

### 同期型
① 各ワーカーが計算が終わるのを待つ<br>
② すべての勾配が出たら、勾配の平均を計算し、親のパラメータを計算する<br>
③ ①に戻る<br>
<img src="./image/3_1_data_parallel_sync.png">


### 非同期型
各ワーカーはお互いの計算を待たず、各子モデルごとに更新を行う。
<img src="./image/3_2_data_parallel_async.png">


現在は同期型の方が精度がいいことが多いので、主流になっている。
<br>
全部自前のサーバーで行い、制御ができる場合→同期型でできる<br>
他人のスマートフォンを使用するなどの分散処理の場合→非同期でやらざるを得ない

## 3-3. GPU
- GPGPU
グラフィック以外の用途で使用されるGPUの総称

- CPU
高性能なコアが小数<br>
複雑で連続的な処理が得意<br>

- GPU
低性能なコアが多数<br>
簡単な並列処理が得意<br>

### GPGPU開発環境
- CUDA
GPU上で並列コンピューティングを行うためのプラットフォーム<br>
NVIDIA社が開発しているGPUのみで使用可能。<br>
Deep Learning用に提供されているので、使いやすい<br>

- OpenCL
ほとんど使われていない<br>
オープンな並列コンピューティングのプラットフォーム<br>
<br>
Deep Learningフレームワーク(Tensorflow, Pytorch)内で実装されているので、使用する際は指定すれば良い

## 3-4. 量子化
64bit浮動小数点を32bitに落とすなどしてメモリ節約と演算処理の削減
### メリット
- 省メモリ化
- 計算の高速化

### デメリット
- 精度の低下

### 計算の高速化
単精度は32bit、倍精度は64bit

### 精度の下限

軽量でありながら複雑な問題に適用できるモデルになる。
