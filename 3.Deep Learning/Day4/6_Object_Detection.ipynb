## IoU : Intersection over Union
物体検出ではクラスラベルだけでなく、物体位置の予測精度も評価したい
$$
IoU = \frac{重なっているエリア}{全エリア}
$$

### Default Boxを多数用意することで生ずる問題
Default Boxを増やすことにより、対象が重複してカウントされる（冗長）<br>
→もっとも正確性の高いものだけ用いる。<br>
物体と背景の検出比率が、物体の方が過剰に検出されたりする。<br>
→PositiveとNegativeの比率を固定する。

いかに解像度を戻す(Up-sampling)するか。<br>
そもそもなぜ、プーリングするのか。<br>

### FCNの基本アイデア

### Deconvolution/Transposed convolution
１. 特徴マップのpixel間隔をstrideだけ空ける<br>
２. 特徴マップのまわりに(kernel size - 1) - paddingだけ余白を作る<br>
３. 畳み込み演算を行う<br>

### 輪郭情報の補完
プーリングによりローカルな情報（輪郭）が失われていく。
低レイヤのプーリング層をよそ事に足し算して、情報を補完する（element-wise addition）

### Upooling
プーリング時にどこが最大の情報を保持していたのか記憶する。
