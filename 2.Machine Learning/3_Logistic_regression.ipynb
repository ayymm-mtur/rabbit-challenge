# 3. ロジスティック回帰
* ある入力からクラスに出力する。入力はベクトルもしくはスカラで、出力は0または1。

* 分類問題のアプローチ
    * 識別的アプローチ
        * $p(C_k | x)$を直接モデル化
    * 生成的アプローチ
        * $p(C_k)$と$p(x |C_k)$をモデル化し、ベイズの定理により$p(C_k | x)$を求める
        $$
        p(C_k | x) = \frac{p(C_k, x)}{p(x)} = \frac{p(x|C_k)p(C_k)}{p(x)}
        $$
    * 識別関数
    
説明変数
$
x = (x_1, x_2, \cdots, x_m)^T \in \mathbb{R}
$
<br>
目的変数
$
y \in {0,1}
$

$\omega^T x$をsigmoid関数に通して、0から1の値に変換する

### sigmoid関数
$$
\sigma(x) = \frac{1}{1+\exp(-x)}
$$

- sigmoid関数の微分はsigmoid関数を用いて表せる
$$
\frac{d\sigma}{dx} = \sigma(x)(1-\sigma(x))
$$

### sigmoid関数の出力をY=1になるよる確率に対応させる
$$
P(Y=1 | x) = \sigma(\omega_0 + \omega_1 x_1 + \cdots + \omega_m x_m)
$$

### 確率分布とパラメータの推定
* ロジスティック回帰モデルでは、ベルヌーイ分布を利用する
    * 確率pで1、p-1で0をとる離散確率分布
    $$
    Y \sim Be(p)\\
    P(y) = p^y(1-p)^{1-y}
    $$
    
* ベルヌーイ分布のパラメータ推定
    * 最尤推定
        * 最尤関数を最大化する
        
        * データは固定し、パラメータを変化させる<br>
        尤度関数
        $$
        \begin{eqnarray}
        P(y_1, y_2, \cdots, y_n;p) &=& \prod_{i=1}^{n}p^{y_i}(1-p)^{1-y_i}\\
        &=& \prod_{i=1}^{n}\sigma(\omega^Tx_i)^{y_i}(1-\omega^Tx_i)^{1-y_i}\\
        &=& L(\omega)
        \end{eqnarray}
        $$
    * 勾配降下法で尤度関数を最大化するパラメータを探す
        $$
        \begin{eqnarray}
        E(\omega) = -\log(L(\omega))
        \end{eqnarray}
        $$
